9. Create a line plot using the x and y values provided below. Label the y-axis as “Y” and
label the x-axis as “X”.
x = [3, 4, 5, 6]
y = [1.5, 2, 2.5, 3]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [3,4,5,6]
y = [1.5,2,2.5,3]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("line plot")
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.plot(x,y , marker='o' , linestyle='-' , color='blue')
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("line plot")
plt.show()
```
